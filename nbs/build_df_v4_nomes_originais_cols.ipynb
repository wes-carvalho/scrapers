```python
from unicodedata import normalize

def pp_text(x):
    '''
    essa função pre-processa o testo, normalizando-o fazendo:
    - remoção de caracteres acidentados;
    - capitalização .upper()
    '''
    
    return normalize('NFKD', x).encode('ASCII','ignore').decode('ASCII').upper()
```

```python
from fuzzywuzzy import fuzz

def suggest(x, marcas):
    
    # threshold de 80
    return marcas[[fuzz.ratio(x, y) > 80 for y in marcas]]
```

```python
def escolhe_marca(df):
    
    marcas = df["marca"].apply(pp_text).unique()

    marca = pp_text(input("Digite a marca desejada: "))

    while marca not in marcas:

        print("\nMarca não encontrada!")

        sugestoes = suggest(marca, marcas)
        if len(sugestoes) > 0:

            print("\nEncontramos algumas marcas similares:\n")

            for i in range(len(sugestoes)):
                print("{} - {}".format(i+1, sugestoes[i]))

            marca = sugestoes[int(input("\nDigite o número correspondente à marca desejada: "))-1]

        else:

            print("\nNenhuma marca similar foi encontrada.")
            print("Segue lista de marcas disponíveis:\n")

            for i in range(len(marcas)):
                print("{} - {}".format(i+1, marcas[i]))

            marca = marcas[int(input("\nDigite o número correspondente à marca desejada: "))-1]

    print("\nMarca registrada!")
    print("="*80)
    print("\n")    
    
    return marca
```

```python
def escolhe_modelo(df):
    
    marca = escolhe_marca(df)
    
    modelos = df[df["marca"]==marca]["modelo"].apply(pp_text).unique()

    modelo = pp_text(input("Digite o modelo desejado: "))

    while modelo not in modelos:

        print("\nmodelo não encontrada!")

        sugestoes = suggest(modelo, modelos)
        if len(sugestoes) > 0:

            print("\nEncontramos alguns modelos similares:\n")

            for i in range(len(sugestoes)):
                print("{} - {}".format(i+1, sugestoes[i]))

            modelo = sugestoes[int(input("\nDigite o número correspondente ao modelo desejado: "))-1]

        else:

            print("\nNenhum modelo similar foi encontrado.")
            print("Segue lista de modelos disponíveis:\n")

            for i in range(len(modelos)):
                print("{} - {}".format(i+1, modelos[i]))

            modelo = modelos[int(input("\nDigite o número correspondente ao modelo desejado: "))-1]

    print("\nmodelo registrado!")
    print("="*80)
    print("\n")
    
    return marca, modelo
```

```python
def escolhe_rg(rg):
    
    try:
        rg.remove("marca")
        rg.remove("modelo")
    except:
        pass
    
    print("Informações disponíveis para o filtro:\n")
    print(*rg, sep="\n")

    resp = pp_text(input("\nVocê deseja utilizar todas as informações para o filtro? (Digite 'sim' ou 'nao'): "))
    while resp not in ["SIM", "NAO"]:
        resp = pp_text(input("\nNão entendi! Por favor, digite somente 'sim' ou 'nao': "))

    if resp == "SIM":
        rg_chosen = rg
    else:
        colunas = input("Digite as informações que você deseja utilizar, separadas por espaço:\n")
        rg_chosen = colunas.split()

        # todo: deixar mais robusto

    return rg_chosen
```

```python
escolhe_rg(rg)
```

```python
import hashlib

def create_hashes(df, rg):
    
    marca, modelo = escolhe_modelo(df)
    
    chosen_rg = escolhe_rg(rg)

    marca_modelo = df[(df["marca"] == marca) & 
                      (df["modelo"] == modelo)][["marca", "modelo"] + chosen_rg].drop_duplicates()
    
    print("\nHashes criadas!")
    print("="*80)
    print("\n")
    
    return marca_modelo.astype(str).agg("-".join, 
                                        axis=1).apply(lambda x: 
                                                      hashlib.sha256(x.encode()).hexdigest()).values, marca_modelo
```

Destas, vamos rodar o DBSCAN pra cada modelo. Com isso, encontraremos outliers.

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
topN_dict = {}

for i in range(std_topN.shape[0]):
    
    rg_data = std_topN.iloc[i][rg].to_dict()
    
    marca = rg_data["marca"]
    modelo = rg_data["modelo"]
    
    print("\n#############################################################")
    print("#############################################################")
    print("\nRG:\n")
    for key, value in rg_data.items():
        print(key, ":", value)
    print("\n")
    print("#############################################################")
    print("#############################################################\n")
    
    aux = std_topN.iloc[[i]][rg].merge(df, on=rg, how="inner").copy()
    
    #################################################################################
#     # visualizando preço vs km com hue de ano
#     f = sns.scatterplot(data=aux, y="preco", x="km", 
#                         hue="ano", palette="rainbow")

#     ax = plt.gca()
#     ax.set_title(marca + " - " + modelo)

#     plt.show()
    #################################################################################

    # clustering

    # selecionando features - apenas numeric & bool
    X = aux.select_dtypes(include=[np.number, np.bool]).copy()

    # selecionando apenas algumas cols - sempre inclua preco!
    cols_features = ["preco", "km"]
    X = aux[cols_features]

    # preenchendo NaNs
    cols_with_nan = X.isnull().any(axis=0)[X.isnull().any(axis=0) == True].index

    for col in cols_with_nan:

        # definindo os atributos pro cruzamento
        cols = ['transmissao', 'categoria', 'blindado', 'cilindrada', 'combustivel']

        # fazendo o agrupamento
        media_col = aux.groupby(cols)[[col]].mean().reset_index()

        # sub-base com valores missing de "col", e apenas as colunas de cruzamento
        X_col_missing = aux.loc[X[col].isnull()][cols]

        # fazendo o cruzamento
        valores_a_preencher = X_col_missing.merge(media_col, 
                                                  on=cols, 
                                                  how="left").set_index(X_col_missing.index)[col]

        # preenchendo vazio com as médias
        X.loc[df[col].isnull(), col] = valores_a_preencher

    # normalização
    X = pd.DataFrame(StandardScaler().fit_transform(X), columns = X.columns, index= X.index)

    #################################################################################
#     # visualizando a normalização
#     f = sns.scatterplot(data=X, y="preco", x="km", palette="rainbow")

#     ax = plt.gca()
#     ax.set_title(marca + " - " + modelo)

#     plt.show()
    #################################################################################

    # modelagem
    db = DBSCAN(eps=0.5, min_samples=2, metric="euclidean").fit(X)

    # cluster labels
    aux["cluster"] = db.labels_

    #################################################################################
    # visualizando preço vs km com hue de cluster
    f = sns.scatterplot(data=aux, y="preco", x="km", 
                        hue="cluster", palette="rainbow")

    ax = plt.gca()
    ax.set_title(marca + " - " + modelo)

    plt.show()
    #################################################################################

    # atualiza o dict com o aux
    topN_dict["|".join([str(x) for x in rg_data.values()])] = aux
        
del aux
```
